```python
def clean_features(df):
    df = pd.get_dummies(df, columns = ["Sex"], dtype = int)
    
    return df

```

```python
def RMSLE(pred_arr, actual_arr):
    count = 0
    
    for pred, actual in zip(pred_arr, actual_arr):
        loss = (np.log(pred+1) - np.log(actual+1))**2
        count += loss
    
    RMSLE = np.sqrt(count / pred_arr.size)
    
    return RMSLE
        
```
